```python
from torch import nn
import torch

class LogisticRegression(nn.Module): 
    def __init__(self, n_features):
        super().__init__()
        self.layer1 = nn.Linear(n_features, 2)
    
    def forward(self, X):
        return self.layer1(X)
```
